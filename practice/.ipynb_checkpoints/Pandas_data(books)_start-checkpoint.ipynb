```python
import pandas as pd
import numpy as np
```

```python
authors = pd.DataFrame({'author_id': [1, 2, 3], 'author_name':['Pushkin', 'Tolstoy', 'Dostoevskiy']}, columns = ['author_id', 'author_name'])
authors
```

```python
books = pd.DataFrame({'author_id':[2, 3, 3, 4], 'book_title':['War and Peace', 'The Idiot', 'Crime and Punishment', 'Fathers and Sons']}, columns = ['author_id', 'book_title'])
books
```

```python
my_library = pd.merge(authors, books, on = 'author_id', how = 'outer')
my_library
```

```python
my_library.loc[my_library['book_title'].isnull(), :]
```

```python
my_library.loc[my_library['author_name'].notnull(), :]
```

```python
my_library['book_title'] = my_library['book_title'].fillna('unknown')
```

```python
my_library[['author_name', 'book_title']] = my_library[['author_name', 'book_title']].fillna('unknown')
my_library
```

```python
my_library.loc[(my_library['author_name'] != 'unknown') & (my_library['book_title'] != 'unknown'),'quantity'] = 1
my_library
```

```python
my_library['quantity'] = my_library['quantity'].fillna(0)
my_library
```

```python
my_library['quantity'] = my_library['quantity'].astype(int)
my_library
```

```python
my_library.set_index('author_id', inplace = True)
```

```python
my_library.reset_index(inplace = True)
my_library
```

```python
my_library = my_library.sort_values(by = 'author_name')
my_library
```

```python
book1 = pd.DataFrame({'author_id': 5, 'author_name':'Den Brown', 'book_title': 'Angels and Daemons', 'quantity': [1]}, columns = ['author_id', 'author_name', 'book_title', 'quantity'] )
book1
```

```python
Mylib2 = pd.concat([my_library, book1], axis = 0, ignore_index = True)
Mylib2
```
