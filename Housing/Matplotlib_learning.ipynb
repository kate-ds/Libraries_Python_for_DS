```python
planets = ['MERCURY', 'VENUS', 'EARTH', 'MOON', 'MARS', 'JUPITER', 'SATURN', 'URANUS', 'NEPTUNE', 'PLUTO']
diameters = [3032, 7521, 7926, 2159, 4221, 88846, 74897, 31763, 30775, 1464]
masses = [0.364, 5.37, 6.58, 0.081, 0.708, 2093, 627, 95.7, 113, 0.0161]
```

```python
rcParams['figure.figsize'] = 9, 12
plt.plot(planets, diameters)
plt.plot(planets, masses)
```

```python
plt.plot(planets, masses)
plt.yscale(value='log')
```

```python
planet_diameter = pd.Series(diameters, index=planets)
planet_diameter
```

```python
planet_diameter.plot(kind='bar', logy=True)
plt.show
```

```python
planet_diameter.plot(kind='barh', logx=True)
```
